# Data Preprocessing

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.25)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier as KNN

# Build the KNN Regressor Model
model = KNN(n_neighbors = 5, metric = 'minkowski', p = 2)
model.fit(train_x,train_y)
y_pred1 = model.predict(test_x)
conf_mat=confusion_matrix(y_pred1,test_y)
print(conf_mat)
print(classification_report(test_y, y_pred1))
```

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB(alpha=1.9)         # alpha by default is 1. alpha must always be > 0. 
# alpha is the '1' in the formula for Laplace Smoothing (P(words))
mnb.fit(train_x,train_y)
y_pred2 = mnb.predict(test_x)
#print("Accuracy Score for Naive Bayes : ", accuracy_score(y_pred1,test_y))
conf_mat=confusion_matrix(y_pred2,test_y)
print(conf_mat)
print(classification_report(test_y, y_pred2))
```

# Support Vector Machines

```python
from sklearn.svm import SVC
svc = SVC(C=1.0,kernel='rbf',gamma='auto')         
# C here is the regularization parameter. Here, L2 penalty is used(default). It is the inverse of the strength of regularization.
# As C increases, model overfits.
# Kernel here is the radial basis function kernel.
# gamma (only used for rbf kernel) : As gamma increases, model overfits.
svc.fit(train_x,train_y)
y_pred3 = svc.predict(test_x)
# print("Accuracy Score for SVC : ", accuracy_score(y_pred2,test_y))
conf_mat=confusion_matrix(y_pred3,test_y)
print(conf_mat)
print(classification_report(test_y, y_pred3))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(train_x,train_y)
y_pred4=lr.predict(test_x)
conf_mat=confusion_matrix(y_pred4,test_y)
print(conf_mat)
print(classification_report(test_y, y_pred4))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state = 20, max_depth = 3) 
model = model.fit(train_x,train_y)
y_pred5 = model.predict(test_x)
conf_mat=confusion_matrix(y_pred5,test_y)
print(conf_mat)
print(classification_report(test_y, y_pred5))
```
